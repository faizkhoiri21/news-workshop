```python
# notebooks/01_eda.ipynb
# --- Setup awal ---
import sys, os
sys.path.append(os.path.abspath("..")) 

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from IPython.display import display

from src.scraping.utils import gold_path

sns.set(style="whitegrid")

# Load data gold
daily = pd.read_parquet(gold_path("daily_articles.parquet"))
quality = pd.read_parquet(gold_path("quality.parquet"))
top_terms = pd.read_parquet(gold_path("top_terms.parquet"))
news = pd.read_parquet(gold_path("news.parquet"))

print("Data shape:", news.shape)
news.head()
```

EDA Dasar

```python
# Agregasi total artikel per sumber
src_counts = daily.groupby("source_normalized")["articles_per_day"].sum().reset_index()

# Plot distribusi artikel per sumber
plt.figure(figsize=(5,6))
ax = sns.barplot(data=src_counts, x="source_normalized", y="articles_per_day", hue="source_normalized", legend=False)
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f", label_type="edge", padding=2)
plt.title("Distribusi Artikel per Sumber (Total)", fontsize=14)
plt.xlabel("Sumber")
plt.ylabel("Jumlah Artikel")
plt.tight_layout()
plt.show()

# Distribusi artikel per hari/per sumber
plt.figure(figsize=(12,6))
ax = sns.lineplot(data=daily, x="date", y="articles_per_day", hue="source_normalized", marker="o")
# pakai locator & formatter biar hanya tanggal
ax.xaxis.set_major_locator(mdates.DayLocator())                  # tick per hari
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))   # format ISO 8601 tanggal saja
# tambahkan label angka di tiap titik
for line in ax.lines:
    x_data = line.get_xdata()
    y_data = line.get_ydata()
    for x, y in zip(x_data, y_data):
        ax.text(x, y, str(int(y)), ha="center", va="bottom", fontsize=12)
plt.title("Distribusi Artikel per Hari")
plt.tight_layout()
plt.show()

# Distribusi panjang artikel (word_count)
plt.figure(figsize=(10,5))
sns.histplot(news["word_count"], bins=50, kde=True)
plt.title("Distribusi Panjang Artikel (word_count)")
plt.xlabel("word_count")
plt.tight_layout()
plt.show()

# Estimasi read time
plt.figure(figsize=(10,5))
ax = sns.histplot(news["read_time_min"], bins=30, kde=False)
# tambahkan angka di atas tiap bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # hanya tampilkan kalau ada isinya
        ax.text(
            p.get_x() + p.get_width()/2,   # posisi tengah bar
            height,                        # posisi y (di atas bar)
            str(int(height)),              # nilai count
            ha="center", va="bottom", fontsize=9
        )
plt.title("Estimasi Read Time (menit)")
plt.xlabel("read_time_min")
plt.tight_layout()
plt.show()

# Missing values per kolom
ax = (news.isna().mean() * 100).plot(kind="bar", figsize=(10,5), title="Missing Value (%) per Kolom")
# tambahkan nilai di atas tiap bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # tampilkan hanya kalau ada missing
        ax.text(
            p.get_x() + p.get_width()/2,  # posisi tengah bar
            height,                       # posisi y
            f"{height:.1f}%",             # format angka 1 desimal + %
            ha="center", va="bottom", fontsize=9
        )
plt.tight_layout()
plt.show()
```

Kebersihan dan Kualitas

```python
# Null rate & dup rate
display(quality)

# Boxplot word_count untuk outlier detection
plt.figure(figsize=(8,5))
sns.boxplot(x=news["word_count"])
plt.title("Outlier Detection: Word Count")
plt.show()

# Rasio artikel tanpa category/author
missing_category = (news["category"].isna().mean() * 100)
missing_author = (news["author"].isna().mean() * 100)
print(f"Rasio artikel tanpa category: {missing_category:.2f}%")
print(f"Rasio artikel tanpa author: {missing_author:.2f}%")
```

Konten dan Topik Ringkas

```python
# Top 15 terms per sumber
top_per_source = (
    top_terms.groupby(["source_normalized", "term"])["count"]
    .sum()
    .reset_index()
    .sort_values(["source_normalized", "count"], ascending=[True, False])
    .groupby("source_normalized")
    .head(15)
)

# Visualisasi: top 15 terms per sumber
g = sns.catplot(
    data=top_per_source,
    x="count", y="term",
    hue="source_normalized", col="source_normalized",
    kind="bar", col_wrap=2, height=5, sharex=False, sharey=False
)
g.set_titles("Top 15 Terms: {col_name}")
g.set_axis_labels("Count", "Term")
plt.show()

# Top 15 terms per kategori
top_per_category = (
    top_terms.groupby(["category", "term"])["count"]
    .sum()
    .reset_index()
    .sort_values(["category", "count"], ascending=[True, False])
    .groupby("category")
    .head(15)
)

# Visualisasi: top 15 terms per kategori
g = sns.catplot(
    data=top_per_category,
    x="count", y="term",
    hue="category", col="category",
    kind="bar", col_wrap=2, height=5, sharex=False, sharey=False
)
g.set_titles("Top 15 Terms: {col_name}")
g.set_axis_labels("Count", "Term")
plt.show()
```

Tren dan Seasonality Ringan

```python
# Rolling mean 7 hari
daily["rolling"] = daily.groupby("source_normalized")["articles_per_day"].transform(lambda s: s.rolling(7,1).mean())

plt.figure(figsize=(12,6))
sns.lineplot(data=daily, x="date", y="articles_per_day", hue="source_normalized", alpha=0.5)
sns.lineplot(data=daily, x="date", y="rolling", hue="source_normalized", linewidth=2)
plt.title("Timeline Artikel Harian + Rolling Mean (7 hari)")
plt.show()

# 5 hari puncak
top5_days = daily.sort_values("articles_per_day", ascending=False).head(5)
print("5 hari puncak penerbitan:")
print(top5_days)

# Lihat kata kunci hari-hari itu
peak_dates = top5_days["date"].unique()
for d in peak_dates:
    print(f"\nKata kunci {d}:")
    print(top_terms.sample(10))
```

Komparasi Antar Sumber

```python
# Perbandingan median word_count antar sumber
plt.figure(figsize=(8,6))
sns.barplot(data=news, x="source_normalized", y="word_count", estimator="median")
plt.title("Median Panjang Artikel per Sumber")
plt.tight_layout()
plt.show()

# Topik dominan berbeda antar sumber (stacked bar)
top10 = top_terms.groupby("term")["count"].sum().nlargest(10).index
pivot = top_terms[top_terms["term"].isin(top10)].pivot_table(
    index="term", columns="source_normalized", values="count", aggfunc="sum", fill_value=0
)

pivot.plot(kind="bar", stacked=True, figsize=(12,6))
plt.title("Topik Dominan Antar Sumber (Stacked)")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```
